```python
import pandas as pd
data = pd.read_csv('howpop_train.csv')
data
```

```python
data.shape
data.head(3).T
```

```python
data.drop(
filter(lambda c: c.endswith("_lognorm"),data.columns),axis=1,inplace=True)#Избавляет от необходимости 
data.describe().T
```

```python
data.describe(include=["object","bool"]).T#Бинарный и категориальные признаки
```

```python
import seaborn as sns
sns.set_style("dark")
sns.set_palette("RdBu")
sns.set_context("notebook",font_scale=1.5,rc={"figure.figsize":(15,5),"axes.titlesize":18})
print(data.published.dtype)
data["published"] = pd.to_datetime(data.published, yearfirst=True)
print(data.published.dtype)
```

```python
data["year"] = [d.year for d in data.published]
data["month"] = [d.month for d in data.published]
data["dayofweek"] = [d.isoweekday() for d in data.published]
data["hour"] = [d.hour for d in data.published]
```

```python
dateFrame=data[data.year>2014].groupby(['year','month'],sort=True)['post_id'].count()
print(dateFrame.plot(kind="bar"))
```

```python
dateFrame
```

```python
dates=data[(data['year']==2015)&(data['month']==3)].copy()
dates=dates.groupby('dayofweek')['post_id'].count().plot(kind="bar",grid=True)
```

```python
dates=data[(data['year']==2015)&(data['month']==3)].copy()
dates['habr']=dates.domain.map(lambda x: 1 if x.startswith('habrahabr.ru') else 0)
dates['geek']=dates.domain.map(lambda x:1 if x.startswith('geektimes.ru') else 0)
dates.groupby('dayofweek')['habr','geek'].sum().plot(kind="bar",grid=True)
```

```python
data.groupby('hour')['comments'].sum().plot(kind="bar")
```

```python
dateFrame=data.copy()
dateFrame=dateFrame.groupby(['author'],sort=True)['post_id'].count()
dateFrame.sort_values(ascending=False, inplace=True)
topAuthor=dateFrame.head(20).index.values
dateFrame=data[data.author.isin(topAuthor)].groupby(['author'],sort=True)['votes_minus'].count()
dateFrame.plot(kind='bar',grid=True, rot=45)
```

```python
dt = data.copy()
dt['geek'] = data.domain.map(lambda x:1 if x.startswith('geektimes.ru') else 0)
select=dt[dt.geek==True].groupby(['hour'],sort=True)['comments'].sum()
select.plot(kind="bar")
```

```python
dt = data.copy()
dt['habr'] = data.domain.map(lambda x:1 if x.startswith('habrahabr.ru') else 0)
select = dt[dt.habr==True].groupby(['hour'],sort=True)['comments'].sum()
select.plot(kind="bar")
```

```python
dateFrame=data.groupby(['hour'],sort=True)['comments'].sum()
dateFrame.plot(kind='bar', rot=90,grid=True, title='Комментарии')
```

```python
data.groupby(['hour'],sort=True)['views'].sum().plot(kind="bar",rot=90,grid=True,title='Просмотры')
```
